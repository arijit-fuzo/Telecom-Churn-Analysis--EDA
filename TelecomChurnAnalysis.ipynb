```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'Senior Citizen': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'Churn Label': ['Churned', 'Not Churned', 'Churned', 'Not Churned', 'Churned', 'Not Churned', 'Churned', 'Not Churned']
})

# Calculate the % of total
data_percent = df.groupby(['Senior Citizen', 'Churn Label']).size().unstack()
data_percent = data_percent.div(data_percent.sum(axis=1), axis=0) * 100  # Convert to percentage

# Plot
data_percent.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(4,4))

# Labels & Titles
plt.title("Churn by Senior Citizen Category (%)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn Label", bbox_to_anchor = (0.9,0.9))
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of your categorical columns
columns = [
    'Offer', 'Phone Service', 'Multiple Lines', 'Internet Service',
    'Internet Type', 'Online Security', 'Online Backup',
    'Device Protection Plan', 'Premium Tech Support',
    'Streaming TV', 'Streaming Movies', 'Streaming Music',
    'Unlimited Data'
]

# Set up the figure: 3 columns per row
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols
plt.figure(figsize=(n_cols * 5, n_rows * 4))

# Create subplots for each feature
for idx, col in enumerate(columns, 1):
    plt.subplot(n_rows, n_cols, idx)
    sns.countplot(data=df, x=col, hue='Churn Label')
    plt.title(f'{col} vs Churn')
    plt.xlabel('')
    plt.ylabel('')
    plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

```python

#📊 Customer Churn Analysis Summary


#This visualization presents how different service features are associated with customer churn behavior. 
#Key insights include:
#- 🧾 Offer Types & Internet Type show noticeable churn variation — especially in Offer E and Fiber Optic users, where churn is higher.
#- 📞 Phone Service and having Multiple Lines have low churn, indicating satisfaction.
#- 🔐 Online Security, Backup, Device Protection, and Premium Tech Support correlate with lower churn when enabled.
#- 📺 Streaming Services (TV, Movies, Music) users churn less, possibly indicating better engagement.
#- 🔄 Unlimited Data users show slightly reduced churn, suggesting higher satisfaction.

#✅ Conclusion:
#Service features like security add-ons and internet type significantly affect churn. Enhancing user experience in these areas can help reduce customer loss.

```
